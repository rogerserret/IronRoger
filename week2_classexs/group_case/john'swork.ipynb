### 3.1 - identifying specified roles useful for our analysis (scientist, engineer or analyst) in the 'position' column

```python
# first we change the strings in the position column to ignore case sensitive alphabelts
df['position'] = df['position'].str.lower()
```

```python
# using the the define() fuction to create a new column named job_position useful for our analysis
def job_position(position):
    if 'scient' in position:
        return 'scientist'
    elif 'analy' in position:
        return 'analyst'
    elif 'engine' in position:
        return 'engineer'
    else:
        return 'others'
    
```

```python
df['job_position'] = df['position'].apply(job_position)
df
```

### 3.2 identifying skills needed  for our analysis in the 'description' column
       - using the pandas.Series.str.contains function to return a new column of boolean values
    Note - flags= re.IGNORECASE to ignore case sensitive alphabelts and
    regex=True to assume the search word is a regular expression.
      -List of skills:

 - SQL
 - Python
 - R
 - Java
 - Predictive Analytics
 - Statistical Modeling
 - Machine learning
 - Deep learning
 - NLP
 - Scala

```python
df['SQL']=df['description'].str.contains('SQL', flags=re.IGNORECASE, regex=True)
df['R']=df['description'].str.contains('R', flags=re.IGNORECASE, regex=True)
df['Predictive_analytics']=df['description'].str.contains('predictive', flags=re.IGNORECASE, regex=True)
df['Statistical_Modeling']=df['description'].str.contains('statistical', flags=re.IGNORECASE, regex=True)
df['Machine_learning']=df['description'].str.contains('machine', flags=re.IGNORECASE, regex=True)
df['Deep_learning']=df['description'].str.contains('deep|learn', flags=re.IGNORECASE, regex=True)
df['NLP']=df['description'].str.contains('nlp', flags=re.IGNORECASE, regex=True)
df['Scala']=df['description'].str.contains('scala', flags=re.IGNORECASE, regex=True)
```
