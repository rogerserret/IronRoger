## NLTK tools to process the text

```python
from nltk.tokenize import word_tokenize
```

```python
mytext = word_tokenize(mytext)
```

```python
mytext
```

```python
len(mytext)
```

```python
# let's put everything in lowercase
lower_text = [word.lower() for word in mytext]
```

```python
# let's count the frequency of words
fdist_v1 = nltk.FreqDist(lower_text)
```

```python
word_freq_v1=dict((word,freq) for word, freq in fdist_v1.items())
```

```python
word_freq_v1
```

```python
sorted(word_freq_v1.items(),key=lambda x:x[1], reverse=True)
#sort it by the value of freq.
```

## Text cleaning (depends on the goal of the analysis)

- remove stop words
- remove puntuation
- lemmatising / stemming
- pronouns, nouns and entities / abbreviations
- prepositions

```python
# remove punctuation spaces numbers - isalpha()
clean_text=[word for word in lower_text if word.isalpha()]
```

```python
len(clean_text)
```

```python
# get stop words - appropiate for text (same language)
from nltk.corpus import stopwords
```

```python
my_stopwords=stopwords.words('english')
```

```python
my_stopwords  #stopwords are words that have no value for the analysis
#if we need to add more to the list we can create our own list
```

```python
clean_text = [word for word in clean_text if word not in my_stopwords]
```

```python
len(clean_text)
```

```python
# let's count the frequency of words - version 2
fdist_v2 = nltk.FreqDist(clean_text)
word_freq_v2=dict((word,freq) for word, freq in fdist_v2.items())
sorted(word_freq_v2.items(),key=lambda x:x[1], reverse=True)
```

```python
porter=nltk.PorterStemmer()

#stemmers reduce words to the stem: horses, horsing, horse to "hors"
```

```python
stem_words=[porter.stem(word) for word in clean_text]
```

```python
# 3rd version of word frequency
fdist_v2 = nltk.FreqDist(stem_words)
word_freq_v2=dict((word,freq) for word, freq in fdist_v2.items())
sorted(word_freq_v2.items(),key=lambda x:x[1], reverse=True)
```

## Get sentiment scores

```python
# textblob - can't deal with nulls
clothingdf = clothingdf.dropna(subset=['Review Text'], how= 'any') 
```

```python
clothingdf.info()
```

```python
clothingdf.iloc[5]['Review Text']
```

```python
example=TextBlob(clothingdf.iloc[5]['Review Text'])
example.sentiment
```

```python
clothingdf.iloc[5]['Review Text']
```

```python
example2=TextBlob(clothingdf.iloc[34]['Review Text'])
example2.sentiment   #it's not a perfect tool, sometimes sentiment doesn't fit reality
```

```python
def get_sentiment_p(text):
    return TextBlob(text).sentiment.polarity
```

```python
clothingdf['sentiment_p']=clothingdf['Review Text'].apply(get_sentiment_p)
```

```python
def get_sentiment_s(text):
    return TextBlob(text).sentiment.subjectivity
```

```python
clothingdf['sentiment_s']=clothingdf['Review Text'].apply(get_sentiment_s)
```

```python
clothingdf.head()
```

## Analysis - avg rating, sentiment, subjectivity per dept div

```python
sent_df = clothingdf.groupby(['Division Name', 'Department Name']).agg({'Rating':'mean', 'sentiment_p':'mean', 'sentiment_s':'mean'})
```

```python
sent_df.columns
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
tweets = pd.read_csv('ever_trump.csv')
```

```python
tweets.head()
```

```python
tweets['created_at'].min()
```

```python
tweets['created_at'].max()
```

```python
tweets.info()
```

```python
tokens = nlp(''.join(str(tweets.text.tolist())))
```

```python
len(tokens)
```

```python
tokens # all words are tokenized in different columns despite we don't see them
```

```python
from collections import Counter
```

```python
items=[x.text for x in tokens.ents]
Counter(items).most_common(20)
```

```python
person_list = []
for ent in tokens.ents:
    if ent.label_=='PERSON':
        person_list.append(ent.text)
```

```python
person_list
```

```python
person_counts=Counter(person_list).most_common(20)
df_person=pd.DataFrame(person_counts, columns=['text', 'count'])
df_person.plot.barh(x='text', y='count');
```
