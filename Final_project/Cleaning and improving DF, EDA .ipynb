```python
# unify columns that belong to the same category:

one_values = ['2','3']

# mezcal + tequila

df['tequila'] = df['tequila'] + df['mezcal']
df['tequila'] = df['tequila'].astype(str)
for i in range(len(Ingredients)):
    if df['tequila'][i] != '0':
        df['tequila'] = df['tequila'].str.replace('2','1')
df['tequila'] = df['tequila'].astype(int)

# champagne + sparkling wine + prosecco

df['sparkling spirits'] = df['champagne'] + df['sparkling wine'] + df['prosecco']
df['sparkling spirits'] = df['sparkling spirits'].astype(str)
for i in range(len(Ingredients)):
    if df['sparkling spirits'][i] != '0':
        df['sparkling spirits'] = df['sparkling spirits'].str.replace('2','1')
        df['sparkling spirits'] = df['sparkling spirits'].str.replace('3','1')
df['sparkling spirits'] = df['sparkling spirits'].astype(int)

# bourbon + whiskey + whisky

df['whisky'] = df['bourbon'] + df['whisky'] + df['whiskey']
df['whisky'] = df['whisky'].astype(str)
for i in range(len(Ingredients)):
    if df['whisky'][i] != '0':
        df['whisky'] = df['whisky'].str.replace('2','1')
        df['whisky'] = df['whisky'].str.replace('3','1')
df['whisky'] = df['whisky'].astype(int)

# maracuya + passion fruit

df['passion fruit'] = df['passion fruit'] + df['maracuya']
df['passion fruit'] = df['passion fruit'].astype(str)
for i in range(len(Ingredients)):
    if df['passion fruit'][i] != '0':
        df['passion fruit'] = df['passion fruit'].str.replace('2','1')
df['passion fruit'] = df['passion fruit'].astype(int)
```
