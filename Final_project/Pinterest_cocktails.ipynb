```python
import pandas as pd
```

# Import data

```python
df = pd.read_csv('pinterest_scraping.csv')
```

```python
df.head()
```

# Clean df

```python
df = df[['createdAt', 'articleDescription']] # we only need these 2 columns
```

```python
df
```

```python
#lower case everything:
df['articleDescription']=df['articleDescription'].str.lower()
```

```python
#drop rows with nulls and reset index

df.dropna(axis=0, inplace = True) 
df.reset_index(drop=True, inplace = True)
```

```python
df.info()
```

```python
df.rename(columns={'createdAt': 'Date',
                   'articleDescription': 'Ingredients'},
          inplace=True, errors='raise')
```

```python
# We only want rows that contain recipes. If a row doesn't containg any of our ingredients we will make the assumption that is not a cocktail recipe:
# Check and remove rows without recipes:

Ingredients = ['mint', 'menth', 'basil', 'cardamon', 'cinnamon', 'coriander', 'ginger', 'lavender', 'rosemar', 'eucalypt',
'lemongrass', 'tarragon', 'vanilla', 'chamomile', 'cocoa', 'milk', 'sparkling water', 'cucumber', 'chilli', 'syrup', 
'coffee', 'bitter', 'cream', 'lemonade', 'lemon', 'banana', 'lime', 'citrus', 'watermelon', 'peach', 'pineapple', 
'orange', 'cranber', 'strawber', 'blackber', 'raspber', 'blueber', 'mango', 'papaya', 'apple', 'passion fruit', 
'apricot', 'maracuya', 'cher', 'kiwi', 'plum', 'grape', 'fig', 'grapefruit', 'pomegranate', 'coconut', 'pumpkin',
'sparkling wine', 'champagne', 'prosecco', 'rum', 'vodka', 'gin', 'vermouth', 'cognac', 'whisky', 'bourbon', 'tequila', 
'cachaça','liqueur', 'mezcal']
```

```python
df["Ingredients"].iloc[17] # example of a row that should be deleted
```

```python
results_set=[]
for t in range(len(df["Ingredients"])):
    check = []
    for i in Ingredients:
        if i in df["Ingredients"].iloc[t]:
            appears = 'ok'
            check.append(appears)
        else:
            appears = 'no'
            check.append(appears) 
    if 'ok' in check:
        results_set.append('keep')
    else:
        results_set.append('drop')
```

```python
df["Check"]=results_set
```

```python
results_set
```

```python
df.info()
```

```python
df=df[df['Check']=='keep']
```

```python
import warnings
warnings.filterwarnings("ignore")
df.reset_index(drop=True, inplace = True)
df.drop(['Check'], axis = 1, inplace = True)
```
