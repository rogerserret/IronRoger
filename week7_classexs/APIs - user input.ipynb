# International space station

```python
# install folium on conda env
# this allows us to follow the movement of the station life
```

```python
response=requests.get('http://api.open-notify.org/iss-now.json')
```

```python
response.json()
```

```python
import folium
```

```python
latlong=[response.json()['iss_position']['latitude'], response.json()['iss_position']['longitude']]
```

```python
mappy=folium.Map(location=latlong, zoom_starts=5)
mappy
```

```python
icon=folium.Icon(color='pink', prefix='fa', icon='rocket', icon_color='black',)
```

```python
icon_data={'location':latlong,'tooltip':'ISS','icon':icon}
```

```python
iss_marker=folium.Marker(**icon_data)
```

```python
iss_marker.add_to(mappy);
```

```python
mappy
```
