# Authentication

```python
import spotipy 
from spotipy.oauth2 import SpotifyClientCredentials
```

# Password protection

```python
import getpass
```

```python
client_id=str(getpass.getpass('enter your client id'))
client_secret=str(getpass.getpass('enter your client secret')) 
```

```python
#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=client_id,
    client_secret=client_secret))
```

# Extract songs ID, name and artist from a single playlist

```python
playlist = sp.user_playlist_tracks("CW", "39sVxPTg7BKwrf2MfgrtcD?si=3003e09975cb4c0a")
```

```python
songs = {}

for track in playlist['tracks']['items']:
    songs[track["track"]["id"]]=track["track"]["name"], track["track"]["artists"][0]["name"]

# we create a dictionary with the id as the key and the name and artist as the values by using a for loop that goes through each song in the list
```

```python
len(songs)
```

# Extract song ID and audio features from a single playlist

```python
audio_features = {}

for track in playlist['tracks']['items']:
    audio_features[track["track"]["id"]] = sp.audio_features(track["track"]["id"])
```
