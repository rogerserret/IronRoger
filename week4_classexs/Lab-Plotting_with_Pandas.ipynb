# Import packages

```python
# Import Packages
import os

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import pandas as pd
import earthpy as et

# Add seaborn general plot specifications
sns.set(font_scale=1.5, style="whitegrid")
```

# Challenge 1: Plot Precipitation Data

```python
df = pd.read_csv('avg-precip-months-seasons.csv') 
```

```python
df.info()
```

```python
df.plot(x='months', y='precip');
```

```python
df.plot(x='months', y='precip', linestyle='dashed', color='green', marker='^');
plt.xlabel('Months');
plt.ylabel('Precipitation');
plt.title('Precipitation vs Months');
```

# Challenge 2: Bar Plot of Precipitation Data

```python
fig, ax = plt.subplots()
ax = fig.add_axes([0,0,1,1])
ax.bar('months', 'precip', data=df, color='green', edgecolor='yellow')
ax.set_xlabel ('Months')
ax.set_ylabel ('Precipitation')
ax.set_title ('Precipitation vs Months')
plt.show()
```

# Challenge 3: Figure with Two Subplots of Precipitation Data

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(8, 10))
plt.suptitle("Overall Figure Title Using the Suptitle Method")
ax1.plot(df.months,
        df.precip,
        color="green",
        linestyle='dashed', marker="o")

ax1.set(ylabel="Mean Precipitation (inches)",
       xlabel="Month",
       title="Plot Challenge 1\nAverage Precipitation Practice Plot")
ax2.bar(df.months,
       df.precip,
       color="green",
       edgecolor="yellow")

ax2.set(ylabel="Mean Precipitation (inches)",
       xlabel="Month",
       title="Bar Plot Challenge 1\nAverage Precipitation Practice Plot")

plt.tight_layout()
plt.show()
# .bar for the plot features and .set for the complement features
```
