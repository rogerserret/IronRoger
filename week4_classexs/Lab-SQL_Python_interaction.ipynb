# Function rentals_may

Create Python function named rentals_may to connect to the Sakila database and return dataframe storing the data from rental table with information for May 2005.

```python
def rentals_may():
    insp = inspect(engine)
    df=pd.read_sql_query('SELECT * FROM rental WHERE DATE_FORMAT(rental_date, "%%m%%y") = "0505"', engine)
    return df
```

```python
df_may = rentals_may()
df_may
```

# Function rentals june

Create Python function named rentals_june to connect to the Sakila database and return dataframe storing the data from rental table with information for June 2005.

```python
def rentals_june():
    insp = inspect(engine)
    df_june = pd.read_sql_query('SELECT * FROM rental WHERE DATE_FORMAT(rental_date, "%%m%%y") = "0605"', engine)
    return df_june
```

```python
df_june = rentals_june()
rentals_june()
```

# Function number of rentals/customer May

Create a Python function to get the number of rentals for each customer in May given the dataframe provided by the function rentals_may. Hint: Consider making use of pandas groupby() and pandas agg()

```python
def rentals_cust_may():
    df_cust_may = df_may.groupby('customer_id').agg({'rental_id':'count'})
    return df_cust_may
```

```python
df_cust_may = rentals_cust_may()
rentals_cust_may()
```

# Function number of rentals/customer June

Create a Python function to get the number of rentals for each customer in June given the dataframe provided by the function rentals_june.

```python
def rentals_cust_june():
    df_cust_june = df_june.groupby('customer_id').agg({'rental_id':'count'})
    return df_cust_june
```

```python
df_cust_june = rentals_cust_june()
rentals_cust_june()
```

# Function check rentals

Write a Python function named check_rentals that given:

- a dataframe with the number of rentals in May by each customer_id
- a dataframe with the number of rentals in June by each customer_id return a new dataframe with the following columns:
 - customer_id
 - Rentals_May
 - Rental_June
 - The difference between the previous columns

Hint: For this part, you can create a join between the two dataframes created before, using the merge function available for pandas dataframes. 

```python
def check_rentals():
    df_rentals = df_cust_may.merge(df_cust_june, how='left', on='customer_id')
    df_rentals['rest'] = df_cust_june['rental_id'] - df_cust_may['rental_id']   
    df_rentals.rename(columns={'rental_id_x': 'count_may','rental_id_y': 'count_june'},inplace=True, errors='raise')
    return df_rentals
```

```python
df_rentals = check_rentals()

df_rentals
```
